# Loading "Sparse" Parquet
Example of loading the parquet files in Python that were generated from sparse arff files:

```python
import pandas
```

```python
df = pandas.read_parquet("dense_pandas.pq")
print(df.dtypes[:3])
df.head()
```

```python
df = pandas.read_parquet("dense_pyarrow.pq")
print(df.dtypes[:3])
df.head()
```

```python
df = pandas.read_parquet("masked_pyarrow.pq")
print(df.dtypes[:3])
df.head()
```
